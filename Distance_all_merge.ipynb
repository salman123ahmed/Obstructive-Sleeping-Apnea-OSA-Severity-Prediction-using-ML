```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [18, 10] 

#import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_timedata = pd.read_csv(r'/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /Grenoble_raw.csv')
UNID = df_timedata['UNID'].unique()
```

```python
len(UNID)
```

```python
Array =[]
for unid in UNID:
  mdict ={}
  mdict["unid"]=unid
  df= pd.read_csv(r'/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /outputs_reults/'+unid+'.csv')
  for index,row in df.iterrows():
    sum = abs(row['SBP']-row['DBP'])+abs(row['DBP']-row['HR'])+abs(row['HR']-row['SBP'])
    #print(index, row['SBP'], row['DBP'],row['HR'], sum)
    mdict[index] = sum
    
  Array.append(mdict)

new_df = pd.DataFrame(Array)
new_df.head()    
```

```python
new_df.to_csv('/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /merge_all_bp.csv')
```
