```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [18, 10] 

from tqdm import tqdm

import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_timedata = pd.read_csv(r'/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /Grenoble_raw.csv')
UNID = df_timedata['UNID'].unique()
len(UNID)
```

```python
sbp ={}
dbp ={}
hr={}

for unid in UNID:
  df= pd.read_csv(r'/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /outputs_reults/'+unid+'.csv')
  sbp[unid]=df['SBP'].to_list()
  dbp[unid]=df['DBP'].to_list()
  hr[unid]=df['HR'].to_list()
  #print(unid, len(sbp[unid]), len(dbp[unid]), len(hr[unid]))

sbp_df = pd.DataFrame(sbp)
dbp_df = pd.DataFrame(dbp)
hr_df = pd.DataFrame(hr)

plt.ioff()
pbar= tqdm(total=39903)
for i in range(len(UNID)):
  for j in range(i+1, len(UNID)):
    pbar.update(1)
    temp_sbp_df=sbp_df[[UNID[i],UNID[j]]]
    temp_dbp_df=dbp_df[[UNID[i],UNID[j]]]
    temp_hr_df=hr_df[[UNID[i],UNID[j]]]

    #print(temp_sbp_df)
    ax=temp_sbp_df.plot()
    bx=temp_dbp_df.plot()
    cx=temp_hr_df.plot()

    fig1 = ax.get_figure()
    fig2 = bx.get_figure()
    fig3 = cx.get_figure()

    fig1.savefig('/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /sbp_plots_all/sbp_'+UNID[i]+"_"+UNID[j]+ '.png')
    fig2.savefig('/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /sbp_plots_all/dbp_'+UNID[i]+"_"+UNID[j]+ '.png')
    fig3.savefig('/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /sbp_plots_all/hr_'+UNID[i]+"_"+UNID[j]+ '.png')
    plt.close(fig1)
    plt.close(fig2)
    plt.close(fig3)
    


```
