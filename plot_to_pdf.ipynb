```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [18, 10] 

from tqdm import tqdm
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df_timedata = pd.read_csv(r'/content/drive/MyDrive/Data Science Projects/OSA Project PhD/Datasets/BP Time series alignment /Grenoble_raw.csv')
UNID = df_timedata['UNID'].unique()
```

```python
from fpdf import FPDF 
from os.path import exists
```

```python

pdf = FPDF()
epw = pdf.w - 2*pdf.l_margin
pdf.set_font('Arial', '', 7.0)
pbar = tqdm(total = (len(UNID)*(len(UNID)-1))/2)
for i in range(len(UNID)):
  for j in range(i, len(UNID)):
    pbar.update(1)
    if not exists('/content/drive/MyDrive/sbp_plots_all/sbp_'+UNID[i]+"_"+UNID[j]+'.png'):
      continue    
    pdf.add_page()
    th = pdf.font_size
    pdf.cell(5)
    pdf.cell(176, 2*th, UNID[i], border=1, align='c')
    pdf.ln(2*th)
    pdf.cell(5)
    pdf.cell(176, 2*th, UNID[j], border=1, align='c')

    pdf.image('/content/drive/MyDrive/sbp_plots_all/sbp_'+UNID[i]+"_"+UNID[j]+'.png', 20,20,150)
pbar.close()
pdf.output('sbp.pdf', 'F')
```
