```python
X = df.iloc[:,1:89].values
Y = df.iloc[:,89].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
             X, Y, test_size = 0.2, random_state=42)
 
knn = KNeighborsClassifier(n_neighbors=7)
 
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_test, y_test))
```
