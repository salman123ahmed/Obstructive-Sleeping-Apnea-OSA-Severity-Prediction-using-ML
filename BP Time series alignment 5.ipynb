```python
def dtw(s, t):
    n, m = len(s), len(t)
    dtw_matrix = np.zeros((n+1, m+1))
    for i in range(n+1):
        for j in range(m+1):
            dtw_matrix[i, j] = np.inf
    dtw_matrix[0, 0] = 0
    
    for i in range(1, n+1):
        for j in range(1, m+1):
            cost = abs(s[i-1] - t[j-1])
            # take last min from a square box
            last_min = np.min([dtw_matrix[i-1, j], dtw_matrix[i, j-1], dtw_matrix[i-1, j-1]])
            dtw_matrix[i, j] = cost + last_min
    return dtw_matrix 
```

```python
from pandas.core.arrays import base
for unid in UNID:
  if unid != '4c51bb1b-8f85-4a20-8c54-a9ec0cab2fc2':
    patient = df_all_patients[df_all_patients['UNID']== unid]
    patient_sbp = patient["SBP"].tolist()
    patient_dbp = patient["DBP"].tolist()
    patient_hr = patient["HR"].tolist()
    #print(dtw(baseline_sbp, patient_sbp))
    #print(dtw(baseline_dbp, patient_dbp))
    #print(dtw(baseline_hr, patient_hr))

    patient_list = []
    for i in range(len(patient_sbp)):
      patient_list.append([patient_sbp[i], patient_dbp[i], patient_hr[i]])

    distance, path = fastdtw(baseline_list, patient_list, dist=euclidean)
    print("Patient:",unid)
    print("Dominating:", "Baseline" if len(baseline_list) > len(patient_list) else "Patient")
    print("Distance:",round(distance, 3))
    print("Path:",path)
    print("--------------------")
```
